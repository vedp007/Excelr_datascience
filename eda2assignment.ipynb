#Encoding

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
categorical_col=df.select_dtypes(include='object').columns
categorical_col
```

```python
ohe_cols = []  # For columns with <= 2 unique values (binary)
le_cols = []   # For columns with > 2 unique values (multi-class)
```

```python
for col in categorical_col:
    if df[col].nunique(dropna=True) <= 2:
        ohe_cols.append(col)
        
    else:
        le_cols.append(col)
```

```python
df_ohe = pd.DataFrame(df[ohe_cols])    # For One-Hot Encoding
df_le = pd.DataFrame(df[le_cols])
```

```python
df_ohe
```

```python
#applying one hot encoding for this dataframe
ohedata=pd.get_dummies(df_ohe,drop_first=True,dtype=int)
ohedata
```

```python
df_le
```

```python
le=LabelEncoder()
df_le_encoded = df_le.copy()
for col in df_le_encoded.columns:
    df_le_encoded[col] = le.fit_transform(df_le_encoded[col])
```

```python
df_le_encoded
```

```python
df
```

```python
df.drop(df[['sex','income']],axis=1,inplace=True)
```

```python
df
```

```python
df=df.join(ohedata,how='right')
```

```python
df
```

```python
df.drop(df[['workclass','education','marital_status','occupation','relationship','race','native_country']],axis=1,inplace=True)
```

```python
df=df.join(df_le_encoded,how='right')
```
