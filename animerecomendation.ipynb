```python
df["genre"]=df["genre"].fillna(df["genre"].mode()[0])
```

```python
df["type"]=df["type"].fillna(df["type"].mode()[0])
```

```python
df["rating"]=df["rating"].fillna(0)
```

```python
df.isnull().sum()
```

```python
pivot=pd.pivot(data=df,index='anime_id',columns='genre',values='rating')
pivot
```

```python
pivot.fillna(0,inplace=True)
pivot
```

```python
#Recomendation system and evaluation
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=cosine_similarity(pivot)
data
```

```python
pd.DataFrame(data)
```

```python
np.fill_diagonal(data,0)
```

```python
similarity = pd.DataFrame(data)
similarity
```

```python
similarity.index = df['anime_id'].unique()
similarity.columns = df['anime_id'].unique()
```

```python
similarity
```

```python
similarity.idxmax()
```

```python
#function that recommends anime
```

```python
anime_map = df.set_index('anime_id')['name'].to_dict()
similarity.index = similarity.index.map(anime_map)
similarity.columns = similarity.columns.map(anime_map)
```

```python
def recommend_anime(title, top_n=5):
    if title not in similarity.columns:
        return "Anime not found in database."
    sim_scores = similarity[title].sort_values(ascending=False).head(top_n)
    return sim_scores
```

```python
recommend_anime("abcdefg",top_n=5)
```

```python
recommend_anime("Naruto",top_n=10)
```

```python
# Can you explain the difference between user-based and item-based collaborative filtering?
```

User-Based Collaborative Filtering: This approach recommends items to a user based on the preferences of similar users. It calculates the similarity between users, often using metrics like cosine similarity, Pearson correlation, or Jaccard similarity.
Item-Based Collaborative Filtering:This approach recommends items based on the similarity between items rather than users. It calculates the similarity between items, typically based on user interactions or ratings. Common metrics include cosine similarity and adjusted cosine similarity.

```python
#2. What is collaborative filtering, and how does it work?
```

Collaborative filtering is a recommendation technique that predicts a user's preferences based on the preferences and behaviors of other users. It operates on the idea that users who have agreed in the past will likely agree in the future.
