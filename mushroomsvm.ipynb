```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('mushroom.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical=df.select_dtypes(include=['object']).columns
```

```python
label=LabelEncoder()
```

```python
for col in categorical:
    if df[col].dtypes=='object':
        df[col]=label.fit_transform(df[col])
```

```python
df
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.hist(bins=20,figsize=(30,25))
```

```python
plt.figure(figsize=(30,25))
df.boxplot()
```

```python
#no outliers present
```

```python
corr=df.corr()
```

```python
plt.figure(figsize=(30,25))
sns.heatmap(corr,annot=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('class',axis=1)
y=df['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
sns.countplot(x='class',data=df)
plt.title('COUNT PLOT')
```

```python
plt.scatter(df['stalk_height'],df['cap_diameter'])
plt.title('SCATTER PLOT')
```

```python
sns.distplot(x)
plt.title('DIST PLOT')
```

```python
from sklearn import svm
from sklearn.svm import SVC
```

```python
model1=SVC(kernel='linear',gamma='scale',C=5)
```

```python
model1.fit(x_train,y_train)
```

```python
ypred=model1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,f1_score
```

```python
print('Accuracy',accuracy_score(y_test,ypred)*100)
```

```python
print(classification_report(y_test,ypred))
```

```python
print('F1 score:',f1_score(y_test,ypred)*100)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'kernel':['rbf'],
    'gamma':['scale'],
    'C':[0.1,4,5]
}
```

```python
grid=GridSearchCV(estimator=model1,param_grid=params)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```
