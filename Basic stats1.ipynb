Standardization of Numerical Variables

```python
numcols=['Volume', 'Avg_Price','Total_Sales_Value','Discount_Rate','Discount_Amount','Net_Sales_Value']
#to print before comparison
print(df[numcols].head)
```

```python
dfstand=df[numcols].copy()
for col in df[numcols]:
    x=df[col].mean()
    std=df[col].std()
    dfstand[col]=(df[col]-x)/std
```

```python
#data after standardization using z-score
dfstand.head
```
